```python
import pandas as pd
from sklearn.datasets import make_moons, make_blobs
import numpy as np
import matplotlib as mpl

from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

```python
cId =pd.read_csv('intermediatefiles/cId.csv',sep=' ')
hId =pd.read_csv('intermediatefiles/haloId.csv',sep=' ')
```

```python
cId
```

```python
datapoints = pd.read_csv('inputfiles/flame.csv',sep='\t',header = None, skiprows = 1)
```

```python
datapoints
```

```python
x=datapoints[1]
y=datapoints[2]
clusters = cId['predicted']
halo = hId['predicted']
```

```python
cmap = plt.cm.jet  # define the colormap
# extract all colors from the .jet map
cmaplist = [cmap(i) for i in range(cmap.N)]
# force the first color entry to be grey
cmaplist[0] = (.5, .5, 1, 1.0)
```

```python
# create the new map
cmap = mpl.colors.LinearSegmentedColormap.from_list(
    'Custom cmap', cmaplist, cmap.N)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].scatter(x,y,c=halo,s=30,cmap=cmap)
ax[1].scatter(x,y,c=clusters,s=30,cmap=cmap)
plt.show()
```
